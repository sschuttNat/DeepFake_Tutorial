```python
import cv2
import numpy as np
import os
```

```python
lomepal_mv = cv2.VideoCapture('videos/lomepal.mp4')
romeo_mv = cv2.VideoCapture('videos/romeo_elvis.mp4')
```

```python
cap = lomepal_mv
#cap = romeo_mv

try:
    if not os.path.exists('data/lomepal'):
        os.makedirs('data/lomepal')
    #if not os.path.exists('data/romeo'):
        #os.makedirs('data/romeo')
except OSError:
    print ('Error: Creating directory of data')
    
```

```python
currentFrame = 0
while(cap.isOpened()):
    # Capture frame-by-frame
    ret, frame = cap.read()
    # Saves image of the current frame in jpg file
    if(currentFrame%100 == 0):
        name = './data/lomepal/frame' + str(currentFrame) + '.jpg'
        #name = './data/romeo/frame' + str(currentFrame) + '.jpg'
        print ('Creating...' + name)
        cv2.imwrite(name, frame)
    # To stop duplicate images
    currentFrame += 1
    if (currentFrame > cap.get(1)):
        # If the number of captured frames is equal to the total number of frames,
        # we stop
        print("Last frame retrieve")
        break
    
# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
```
